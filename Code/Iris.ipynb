- we can differentiate between Iris-setosa and others, but its slightly difficult to differentiate between Iris-virginica and Iris-versicolor 

## Iris-setosa
- Iris-setosa has a small lenght of sepal (<= 6 cm) and most of them have large width (>= 3 cm)

- The length (<=2 cm) and width (<= 1 cm) of petal is very small in Iris-setosa


## Iris-virginica
- The length and width of petal is sightly large in Iris virginica
- range of width is between(1.4 - 2.5) cm
- range of length is between(4.5 - 7) cm


## Iris-versicolor
- The length and width of petal is large in Iris versicolor
- range of width is between(1 - 1.8) cm
- range of length is between(3 - 5) cm

```python
from pandas import *
from numpy import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=read_csv('Iris.csv')
data.head()
```

```python
data.info()
```

```python
data.columns=data.columns.str.lower()
data.head()
```

```python
data['species'].unique()
```

#    

#       

```python
sns.set_context('talk')
ax=sns.countplot(x='species',data=data)
total=float(len(data))
for i in ax.patches:
    ax.text(i.get_x()+i.get_width()/2,i.get_height()+3,'{:.0f}%'.format((i.get_height()*100)/total))
```

### We have the same numbers of species' types

#       

```python
sns.set_context('talk')
sns.relplot(x='sepallengthcm',y='sepalwidthcm',hue='species',data=data,kind='scatter',style='species')
```

```python
sns.set_context('talk')
sns.relplot(x='sepallengthcm',y='sepalwidthcm',col='species',data=data,kind='scatter')
```

### From this plot , We can see that :
-Iris setosa has a small lenght of sepal and large width ( <= 6)

-but we can not differentiate between Iris versicolor & Iris virginica

```python
sns.set_context('talk')
sns.relplot(x='petallengthcm',y='petalwidthcm',data=data,hue='species')
```

```python
sns.set_context('talk')
sns.relplot(x='petallengthcm',y='petalwidthcm',data=data,col='species')
```
